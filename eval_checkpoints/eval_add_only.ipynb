```python
import torch
import os
import matplotlib.pyplot as plt

checkpoint_paths = ['../test_out/out-check-add-only-plain', '../test_out/out-check-add-only-eval']
nums = [0, 2000, 4000, 6000, 8000, 10000]  # 在开始添加一个零点

device = 'cuda' if torch.cuda.is_available() else 'cpu'

# 存储每个检查点的最佳精度
best_accuracies = {path: [0] for path in checkpoint_paths}  # 在开始添加一个零点

for path in checkpoint_paths:
    for num in nums[1:]:  # 从第二个元素开始循环
        checkpoint_name = f'ckpt_{num}_acc.pt'
        ckpt_path = os.path.join(path, checkpoint_name)
        checkpoint = torch.load(ckpt_path, map_location=device)
        best_accuracies[path].append(checkpoint['best_accuracy'])  # 将精度转换为百分比

# 创建图表
plt.figure(figsize=(10, 6))

colors = ['#1f77b4', '#d62728']
labels = ['Plain', 'Eval Format']

for path, color, label in zip(checkpoint_paths, colors, labels):
    plt.plot(nums, best_accuracies[path], marker='o', color=color, label=label, linewidth=3)

# 设置标题和标签
plt.xlabel('Number of train examples', fontsize=20)
plt.ylabel('Best Accuracy (%)', fontsize=20)

# 设置图例
plt.legend(fontsize=18, frameon=True, shadow=True, borderpad=1, loc='lower right')

# 设置网格线
plt.grid(True)

# 设置刻度字体大小
plt.xticks(fontsize=14)
plt.yticks(fontsize=12)

# 显示图表
plt.show()
```

```python
# 打印输出所有点的acc
for path in checkpoint_paths:
    print(f'Path: {path}')
    for num, acc in zip(nums, best_accuracies[path]):
        print(f'Num: {num}, Acc: {acc}')
    print()
```

```python
import torch
import os
import matplotlib.pyplot as plt

checkpoint_paths = ['../test_out/out-check-judge-only-W1', '../test_out/out-check-judge-only-W2', '../test_out/out-check-mixed-eval-L+W1', '../test_out/out-check-mixed-eval-L+W2']
nums = [0, 2000, 4000, 6000, 8000, 10000]  # 在开始添加一个零点

device = 'cuda' if torch.cuda.is_available() else 'cpu'

# 存储每个检查点的最佳精度
best_judge_accuracies = {path: [0] for path in checkpoint_paths}  # 在开始添加一个零点

for path in checkpoint_paths:
    for num in nums[1:]:  # 从第二个元素开始循环
        if path == '../test_out/out-check-judge-only-W1':
            checkpoint_name = f'ckpt_judge_only_{num}_W1_judge_acc.pt'
        elif path == '../test_out/out-check-judge-only-W2':
            checkpoint_name = f'ckpt_judge_only_{num}_W2_judge_acc.pt'
        elif path == '../test_out/out-check-mixed-eval-L+W1':
            checkpoint_name = f'ckpt_mixed_L+W1_{num}_judge_acc.pt'
        elif path == '../test_out/out-check-mixed-eval-L+W2':
            checkpoint_name = f'ckpt_mixed_L+W2_{num}_judge_acc.pt'
        ckpt_path = os.path.join(path, checkpoint_name)
        checkpoint = torch.load(ckpt_path, map_location=device)
        best_judge_accuracies[path].append(checkpoint['best_judgeacc'])  # 将精度转换为百分比

# 创建图表
plt.figure(figsize=(10, 6))

colors = ['#1f77b4', '#d62728', '#2ca02c', '#ff7f0e']
labels = ['Judge Only(W1)', 'Judge Only(W2)', 'Mixed L+W1(Best Judgment Model)', 'Mixed L+W2(Best Judgment Model)']
linestyles = ['-', '--', '-', '--']  # 设置线的样式

for path, color, label, linestyle in zip(checkpoint_paths, colors, labels, linestyles):
    plt.plot(nums, best_judge_accuracies[path], marker='o', color=color, label=label, linewidth=3, linestyle=linestyle)

# 设置标题和标签
plt.xlabel('Number of addition train examples', fontsize=20)
plt.ylabel('Best Judgment Accuracy (%)', fontsize=20)

# 设置图例
plt.legend(fontsize=18, frameon=True, shadow=True, borderpad=1, loc='lower right')

# 设置网格线
plt.grid(True)

# 设置刻度字体大小
plt.xticks(fontsize=14)
plt.yticks(fontsize=12)

# 显示图表
plt.show()
```
