## 加载judge-only模型和addition from judge模型，对所有参数做平均

```python
from main_utils import *
from model import GPTConfig, GPT
import torch
from contextlib import nullcontext
import re

pre_model_path = 'out/out-judge-only/ckpt_judge_acc.pt'
# init from a model saved in a specific directory
pre_checkpoint = torch.load(pre_model_path, map_location='cuda')
pre_state_dict = pre_checkpoint['model']
unwanted_prefix = '_orig_mod.'
for k,v in list(pre_state_dict.items()):
    if k.startswith(unwanted_prefix):
        pre_state_dict[k[len(unwanted_prefix):]] = pre_state_dict.pop(k)
```

```python
fine_model_path = 'out/out-addition-from-judge/ckpt_acc.pt'
# init from a model saved in a specific directory
fine_checkpoint = torch.load(fine_model_path, map_location='cuda')
fine_state_dict = fine_checkpoint['model']
unwanted_prefix = '_orig_mod.'
for k,v in list(fine_state_dict.items()):
    if k.startswith(unwanted_prefix):
        fine_state_dict[k[len(unwanted_prefix):]] = fine_state_dict.pop(k)
```

```python
state_dict = {}
for key in fine_state_dict.keys():
    state_dict[key] = (fine_state_dict[key] + pre_state_dict[key]) / 2
```

```python
from test_utils import model_tester

device = 'cuda'
tester = model_tester(
    model_path=fine_model_path,
    model_state=state_dict,
    extra_test_file='./test_data/extra_num_judge_prompt.txt',
    add_noise_test_file='./test_data/add_noise_judge_prompt.txt',
    mode='judge_op',
    mydevice=device
)
```

```python
tester.test_add_noise()
```

```python
tester.test_extra()
```

即使做了平均，还是“失去”了判断能力。下面查看其平均之后的计算能力

```python
from test_utils import model_addition_tester

device = 'cuda'
tester = model_addition_tester(
    model_path=fine_model_path,
    model_state=state_dict,
    test_file='./data/bal/test_3digit_10000.txt',
    mydevice=device,
    data_format='eval_format'
)
```

```python
tester.test_addition()
```

计算准确率整体只有30%。
